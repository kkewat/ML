```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading the dataset

```python
data = pd.read_csv('/content/data_cleaned.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

Seperating features and targets

```python
X = data.drop(['Survived'],axis=1)
y = data['Survived']
```

creating train and validation set

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=45, stratify=y, test_size=0.25)
```

```python
y_train.value_counts(normalize=True)
```

```python
X_train.shape, y_train.shape
```

Import decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(random_state=10)
```

```python
#fitting the model
dt_model.fit(X_train,y_train)
```

```python
#checking the training score
dt_model.score(X_train, y_train)
```

```python
#Checking the score for Validation set
dt_model.score(X_valid, y_valid)
```

```python
dt_model.predict(X_valid)
```

```python
dt_model.predict_proba(X_valid)
```

```python
y_pred = dt_model.predict_proba(X_valid)[:,1]
```

```python
new_y = []
for i in range (len(y_pred)):
  if y_pred[i] < 0.6:
    new_y.append(0)
  else:
    new_y.append(1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_valid, new_y)
```

Changing the max depth

```python
train_accuracy = []
validation_accuracy = []
for depth in range(1,10):
  dt_model = DecisionTreeClassifier(max_depth = depth, random_state = 10)
  dt_model.fit(X_train, y_train)
  train_accuracy.append(dt_model.score(X_train, y_train))
  validation_accuracy.append(dt_model.score(X_valid, y_valid))
```

```python
frame = pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'validation_acc':validation_accuracy})
frame.head()
```

```python
# Plotting these dataframe
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'],frame['train_acc'], marker='o')
plt.plot(frame['max_depth'],frame['validation_acc'], marker='o')
plt.xlabel('Depth of tree')
plt.ylabel('Accuracy')
plt.legend(['train_acc','validation_acc'])
```

```python
dt_model = DecisionTreeClassifier(max_depth=4, max_leaf_nodes=25, random_state=10)
```

```python
dt_model.fit(X_train,y_train)
dt_model.score(X_train,y_train)
```

```python
dt_model.score(X_valid,y_valid)
```

```python
from sklearn import tree
```
